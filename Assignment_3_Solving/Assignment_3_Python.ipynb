```python
# display boxplots
plt.subplots(figsize = (6,6))
plt.subplot(121)   #represents the number of rows in the grid, the number of columns in the grid and the position of the subplot in the grid 
plt.boxplot(data['Unit A'])
plt.title('Unit A')
plt.subplot(122)
plt.boxplot(data['Unit B'])
plt.title('Unit B')
plt.show()
```

```python
#  side-by-side distribution plot
plt.figure(figsize = (8,6))
labels = ['Unit A', 'Unit B']
sns.distplot(data['Unit A'], kde = True)
sns.distplot(data['Unit B'],hist = True)
plt.legend(labels)
plt.show()
```

```python
# quantile-quantile (Q-Q) plots
sm.qqplot(data["Unit A"], line = 'q')
plt.title('Unit A')
sm.qqplot(data["Unit B"], line = 'q')
plt.title('Unit B')
plt.show()
```

```python
statistic , p_value = stats.ttest_ind(data['Unit A'],data['Unit B'], alternative = 'two-sided')
print(f'statistic={statistic:.4f}, p_value= {p_value:.4f}')
```

```python
alpha = 0.05/2
print('Significnace=%.4f, p_value=%.4f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis there is a significance difference between two Units A and B')
```

table = [[50,142,131,70],

 H0= All proportions are equal

```python
data=pd.read_csv(r"BuyerRatio.csv")
data
```

```python
print('\n Describe:');   print(data.describe())
print('\n Info:');   print(data.info())
print('\n Null values:');   print(data.isnull().sum())
print('\n Duplicated:');   print(data[data.duplicated()])
```

```python
table = [[50,142,131,70],
        [435,1523,1356,750]]
print(table)
```

```python
stat, p, dof, exp = stats.chi2_contingency(table) 
print(stat,"\n", p,"\n", dof,"\n", exp)
```

```python
observed = np.array(table).flatten();  print('observed: \n',observed)
expected = exp.flatten();   print('expected: \n',expected)
```

```python
# Compare Evidences with Hypothesis using chisquare
statistics, p_value = stats.chisquare(observed, expected, ddof = 3)
print(f'statistic={statistics:.4f}, p_value= {p_value}')
```

```python
alpha = 0.05
print('Significnace=%.4f, p_value=%.4f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between TAT of reports of the laboratories')
else:
    print('We fail to reject Null hypothesis there is a significance difference between TAT of reports of the laboratories')
```

 ##    Q4. TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

Minitab File: CustomerOrderForm.mtw
 
     


```python
data=pd.read_csv(r"Costomer+OrderForm.csv")
data.head()
```

```python
print('\n Describe:');   print(data.describe())
print('\n Info:');   print(data.info())
print('\n Null values:');   print(data.isnull().sum())
# print('\n Duplicated:');   print(data[data.duplicated()])
```

```python
error_free_counts=[]
for i in range(0, 4):
    error_free_count = data[data.iloc[:, i] == 'Error Free'].shape[0]
    error_free_counts.append(error_free_count)
print(error_free_counts)
```

```python
contingency_table=[]
for i in range(0, 4):
    Error_free_count = (data.iloc[:, i] == 'Error Free').sum()
    Defective_count = (data.iloc[:, i] == 'Defective').sum()
    inner_list = [Error_free_count, Defective_count]
    contingency_table.append(inner_list)
    
# Transpose the contingency_table
contingency_table = list(map(list, zip(*contingency_table)))
print(contingency_table)
```

```python
stat, p, df, exp = stats.chi2_contingency(contingency_table)
print("Statistics = ",stat,"\n",'P_Value = ', p,'\n', 'degree of freedom =', df,'\n', 'Expected Values = ', exp)
```

```python
observed = np.array(contingency_table).flatten();  print('observed: \n',observed)
expected = exp.flatten();   print('expected: \n',expected)
```

```python
# Compare Evidences with Hypothesis using chisquare
test_statistic , p_value = stats.chisquare(observed, expected, ddof = df)
print("Test Statistic = ",test_statistic,'\n', 'p_value =',p_value)
```

```python
alpha = 0.05
print('Significnace=%.4f, p_value=%.4f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between TAT of reports of the laboratories')
else:
    print('We fail to reject Null hypothesis there is a significance difference between TAT of reports of the laboratories')
```
